```python
# First, data preparation
# 18 female, 14 male in outer
# 16 female, 14 male in inner
female = ["00123.obj","00129.obj","00136.obj","00137.obj","00140.obj", "00147.obj","00148.obj","00151.obj","00152.obj","00156.obj","00163.obj","00167.obj","00170.obj","00176.obj","00185.obj","00187.obj","00190.obj","00191.obj"]
male = ["00122.obj","00127.obj","00134.obj","00135.obj","00149.obj","00154.obj","00160.obj","00168.obj","00169.obj","00174.obj","00175.obj","00179.obj","00180.obj","00188.obj"]

import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Remove .obj extension from IDs for matching
female_ids = [int(id.replace('.obj', '')) for id in female]
male_ids = [int(id.replace('.obj', '')) for id in male]

# Load only the detailed measurement files
inner_distance = pd.read_csv("/Users/paulinagerchuk/Downloads/dataset-segment-analyse/obj_4ddress_labelled_files/analysis_results/detailed_distance_measurements_inner.csv")
outer_distance = pd.read_csv("/Users/paulinagerchuk/Downloads/dataset-segment-analyse/obj_4ddress_labelled_files/analysis_results/detailed_distance_measurements.csv")

inner_chamfer = pd.read_csv("/Users/paulinagerchuk/Downloads/dataset-segment-analyse/obj_4ddress_labelled_files/analysis_results/detailed_chamfer_measurements.csv") 
outer_chamfer = pd.read_csv("/Users/paulinagerchuk/Downloads/dataset-segment-analyse/obj_4ddress_labelled_files/analysis_results/detailed_chamfer_measurements_outer.csv")

# Filter for overall measurements only
inner_distance_overall = inner_distance[inner_distance['body_part'] == 'overall']
outer_distance_overall = outer_distance[outer_distance['body_part'] == 'overall']
inner_chamfer_overall = inner_chamfer[inner_chamfer['body_part'] == 'overall']
outer_chamfer_overall = outer_chamfer[outer_chamfer['body_part'] == 'overall']

# Calculate statistics for distance measurements
print("\n=== DETAILED DISTANCE MEASUREMENTS ===")

# Inner dataset
inner_female_dist = inner_distance_overall[inner_distance_overall['model_id'].isin(female_ids)]['mean'].mean()
inner_male_dist = inner_distance_overall[inner_distance_overall['model_id'].isin(male_ids)]['mean'].mean()
inner_female_dist_std = inner_distance_overall[inner_distance_overall['model_id'].isin(female_ids)]['mean'].std()
inner_male_dist_std = inner_distance_overall[inner_distance_overall['model_id'].isin(male_ids)]['mean'].std()

# Outer dataset
outer_female_dist = outer_distance_overall[outer_distance_overall['model_id'].isin(female_ids)]['mean'].mean()
outer_male_dist = outer_distance_overall[outer_distance_overall['model_id'].isin(male_ids)]['mean'].mean() 
outer_female_dist_std = outer_distance_overall[outer_distance_overall['model_id'].isin(female_ids)]['mean'].std()
outer_male_dist_std = outer_distance_overall[outer_distance_overall['model_id'].isin(male_ids)]['mean'].std()

print("\nInner Dataset Distance Results:")
print(f"Female mean: {inner_female_dist:.6f}, std: {inner_female_dist_std:.6f}")
print(f"Male mean: {inner_male_dist:.6f}, std: {inner_male_dist_std:.6f}")
print(f"Difference (Female - Male): {inner_female_dist - inner_male_dist:.6f}")

print("\nOuter Dataset Distance Results:")
print(f"Female mean: {outer_female_dist:.6f}, std: {outer_female_dist_std:.6f}")
print(f"Male mean: {outer_male_dist:.6f}, std: {outer_male_dist_std:.6f}")
print(f"Difference (Female - Male): {outer_female_dist - outer_male_dist:.6f}")

# Calculate statistics for chamfer measurements
print("\n=== DETAILED CHAMFER MEASUREMENTS ===")

# Inner dataset
inner_female_chamfer = inner_chamfer_overall[inner_chamfer_overall['model_id'].isin(female_ids)]['chamfer_distance'].mean()
inner_male_chamfer = inner_chamfer_overall[inner_chamfer_overall['model_id'].isin(male_ids)]['chamfer_distance'].mean()
inner_female_chamfer_std = inner_chamfer_overall[inner_chamfer_overall['model_id'].isin(female_ids)]['chamfer_distance'].std()
inner_male_chamfer_std = inner_chamfer_overall[inner_chamfer_overall['model_id'].isin(male_ids)]['chamfer_distance'].std()

# Outer dataset
outer_female_chamfer = outer_chamfer_overall[outer_chamfer_overall['model_id'].isin(female_ids)]['chamfer_distance'].mean()
outer_male_chamfer = outer_chamfer_overall[outer_chamfer_overall['model_id'].isin(male_ids)]['chamfer_distance'].mean()
outer_female_chamfer_std = outer_chamfer_overall[outer_chamfer_overall['model_id'].isin(female_ids)]['chamfer_distance'].std()
outer_male_chamfer_std = outer_chamfer_overall[outer_chamfer_overall['model_id'].isin(male_ids)]['chamfer_distance'].std()

print("\nInner Dataset Chamfer Results:")
print(f"Female mean: {inner_female_chamfer:.6f}, std: {inner_female_chamfer_std:.6f}")
print(f"Male mean: {inner_male_chamfer:.6f}, std: {inner_male_chamfer_std:.6f}")
print(f"Difference (Female - Male): {inner_female_chamfer - inner_male_chamfer:.6f}")

print("\nOuter Dataset Chamfer Results:")
print(f"Female mean: {outer_female_chamfer:.6f}, std: {outer_female_chamfer_std:.6f}")
print(f"Male mean: {outer_male_chamfer:.6f}, std: {outer_male_chamfer_std:.6f}")
print(f"Difference (Female - Male): {outer_female_chamfer - outer_male_chamfer:.6f}")

# Statistical testing function
def create_visualizations(dataset_name, female_p2s, male_p2s, female_chamfer, male_chamfer):
    # Create dataframe for easier plotting
    data_p2s = pd.DataFrame({
        'Distance (cm)': pd.concat([female_p2s, male_p2s]),
        'Gender': ['Female'] * len(female_p2s) + ['Male'] * len(male_p2s)
    })
    
    data_chamfer = pd.DataFrame({
        'Distance (cm)': pd.concat([female_chamfer, male_chamfer]),
        'Gender': ['Female'] * len(female_chamfer) + ['Male'] * len(male_chamfer)
    })
    
    # Box plots
    fig, axes = plt.subplots(1, 2, figsize=(15, 6))
    
    sns.boxplot(x='Gender', y='Distance (cm)', data=data_p2s, ax=axes[0])
    axes[0].set_title(f'P2S Distance by Gender ({dataset_name} Dataset)')
    
    sns.boxplot(x='Gender', y='Distance (cm)', data=data_chamfer, ax=axes[1])
    axes[1].set_title(f'Chamfer Distance by Gender ({dataset_name} Dataset)')
    
    plt.tight_layout()
    plt.show()
    
    # QQ Plots for normality check
    fig, axes = plt.subplots(2, 2, figsize=(15, 12))
    
    # P2S Female
    stats.probplot(female_p2s, dist="norm", plot=axes[0, 0])
    axes[0, 0].set_title(f'{dataset_name} P2S Distance - Female Q-Q Plot')
    
    # P2S Male
    stats.probplot(male_p2s, dist="norm", plot=axes[0, 1])
    axes[0, 1].set_title(f'{dataset_name} P2S Distance - Male Q-Q Plot')
    
    # Chamfer Female
    stats.probplot(female_chamfer, dist="norm", plot=axes[1, 0])
    axes[1, 0].set_title(f'{dataset_name} Chamfer Distance - Female Q-Q Plot')
    
    # Chamfer Male
    stats.probplot(male_chamfer, dist="norm", plot=axes[1, 1])
    axes[1, 1].set_title(f'{dataset_name} Chamfer Distance - Male Q-Q Plot')
    
    plt.tight_layout()
    plt.show()
    
    # Histograms
    fig, axes = plt.subplots(2, 2, figsize=(15, 12))
    
    sns.histplot(female_p2s, kde=True, ax=axes[0, 0])
    axes[0, 0].set_title(f'{dataset_name} P2S Distance - Female')
    
    sns.histplot(male_p2s, kde=True, ax=axes[0, 1])
    axes[0, 1].set_title(f'{dataset_name} P2S Distance - Male')
    
    sns.histplot(female_chamfer, kde=True, ax=axes[1, 0])
    axes[1, 0].set_title(f'{dataset_name} Chamfer Distance - Female')
    
    sns.histplot(male_chamfer, kde=True, ax=axes[1, 1])
    axes[1, 1].set_title(f'{dataset_name} Chamfer Distance - Male')
    
    plt.tight_layout()
    plt.show()

# Statistical testing function 
def test_gender_differences():
    # Extract actual data points for inner dataset (P2S distance)
    inner_female_dist_values = inner_distance_overall[inner_distance_overall['model_id'].isin(female_ids)]['mean']
    inner_male_dist_values = inner_distance_overall[inner_distance_overall['model_id'].isin(male_ids)]['mean']
    
    # Extract actual data points for inner dataset (Chamfer distance)
    inner_female_chamfer_values = inner_chamfer_overall[inner_chamfer_overall['model_id'].isin(female_ids)]['chamfer_distance']
    inner_male_chamfer_values = inner_chamfer_overall[inner_chamfer_overall['model_id'].isin(male_ids)]['chamfer_distance']
    
    # Extract actual data points for outer dataset (P2S distance)
    outer_female_dist_values = outer_distance_overall[outer_distance_overall['model_id'].isin(female_ids)]['mean']
    outer_male_dist_values = outer_distance_overall[outer_distance_overall['model_id'].isin(male_ids)]['mean']
    
    # Extract actual data points for outer dataset (Chamfer distance)
    outer_female_chamfer_values = outer_chamfer_overall[outer_chamfer_overall['model_id'].isin(female_ids)]['chamfer_distance']
    outer_male_chamfer_values = outer_chamfer_overall[outer_chamfer_overall['model_id'].isin(male_ids)]['chamfer_distance']
    
    # Perform statistical tests
    print("\n=== STATISTICAL SIGNIFICANCE TESTS ===")
    
    # Inner Dataset Tests
    print("\n--- INNER DATASET ---")
    
    # 1. T-tests (parametric) for inner dataset
    inner_p2s_ttest = stats.ttest_ind(inner_female_dist_values, inner_male_dist_values, equal_var=False)
    inner_chamfer_ttest = stats.ttest_ind(inner_female_chamfer_values, inner_male_chamfer_values, equal_var=False)
    
    print("\nT-test results (Inner Dataset P2S Distance):")
    print(f"t-statistic: {inner_p2s_ttest.statistic:.4f}")
    print(f"p-value: {inner_p2s_ttest.pvalue:.4f}")
    print(f"Statistically significant (p < 0.05): {inner_p2s_ttest.pvalue < 0.05}")
    
    print("\nT-test results (Inner Dataset Chamfer Distance):")
    print(f"t-statistic: {inner_chamfer_ttest.statistic:.4f}")
    print(f"p-value: {inner_chamfer_ttest.pvalue:.4f}")
    print(f"Statistically significant (p < 0.05): {inner_chamfer_ttest.pvalue < 0.05}")
    
    # 2. Mann-Whitney U tests (non-parametric) for inner dataset
    inner_p2s_mw = stats.mannwhitneyu(inner_female_dist_values, inner_male_dist_values)
    inner_chamfer_mw = stats.mannwhitneyu(inner_female_chamfer_values, inner_male_chamfer_values)
    
    print("\nMann-Whitney U test results (Inner Dataset P2S Distance):")
    print(f"U-statistic: {inner_p2s_mw.statistic:.4f}")
    print(f"p-value: {inner_p2s_mw.pvalue:.4f}")
    print(f"Statistically significant (p < 0.05): {inner_p2s_mw.pvalue < 0.05}")
    
    print("\nMann-Whitney U test results (Inner Dataset Chamfer Distance):")
    print(f"U-statistic: {inner_chamfer_mw.statistic:.4f}")
    print(f"p-value: {inner_chamfer_mw.pvalue:.4f}")
    print(f"Statistically significant (p < 0.05): {inner_chamfer_mw.pvalue < 0.05}")
    
    # Check normality assumptions for inner dataset
    print("\n=== NORMALITY TESTS (INNER DATASET) ===")
    _, p_inner_female_p2s = stats.shapiro(inner_female_dist_values)
    _, p_inner_male_p2s = stats.shapiro(inner_male_dist_values)
    _, p_inner_female_chamfer = stats.shapiro(inner_female_chamfer_values)
    _, p_inner_male_chamfer = stats.shapiro(inner_male_chamfer_values)
    
    print(f"Shapiro-Wilk test p-values (p < 0.05 indicates non-normal distribution):")
    print(f"Inner Female P2S: {p_inner_female_p2s:.4f} (Normal: {p_inner_female_p2s >= 0.05})")
    print(f"Inner Male P2S: {p_inner_male_p2s:.4f} (Normal: {p_inner_male_p2s >= 0.05})")
    print(f"Inner Female Chamfer: {p_inner_female_chamfer:.4f} (Normal: {p_inner_female_chamfer >= 0.05})")
    print(f"Inner Male Chamfer: {p_inner_male_chamfer:.4f} (Normal: {p_inner_male_chamfer >= 0.05})")
    
    # Create visualizations for inner dataset
    create_visualizations("Inner", inner_female_dist_values, inner_male_dist_values, 
                         inner_female_chamfer_values, inner_male_chamfer_values)
    
    # Outer Dataset Tests
    print("\n--- OUTER DATASET ---")
    
    # 1. T-tests (parametric) for outer dataset
    outer_p2s_ttest = stats.ttest_ind(outer_female_dist_values, outer_male_dist_values, equal_var=False)
    outer_chamfer_ttest = stats.ttest_ind(outer_female_chamfer_values, outer_male_chamfer_values, equal_var=False)
    
    print("\nT-test results (Outer Dataset P2S Distance):")
    print(f"t-statistic: {outer_p2s_ttest.statistic:.4f}")
    print(f"p-value: {outer_p2s_ttest.pvalue:.4f}")
    print(f"Statistically significant (p < 0.05): {outer_p2s_ttest.pvalue < 0.05}")
    
    print("\nT-test results (Outer Dataset Chamfer Distance):")
    print(f"t-statistic: {outer_chamfer_ttest.statistic:.4f}")
    print(f"p-value: {outer_chamfer_ttest.pvalue:.4f}")
    print(f"Statistically significant (p < 0.05): {outer_chamfer_ttest.pvalue < 0.05}")
    
    # 2. Mann-Whitney U tests (non-parametric) for outer dataset
    outer_p2s_mw = stats.mannwhitneyu(outer_female_dist_values, outer_male_dist_values)
    outer_chamfer_mw = stats.mannwhitneyu(outer_female_chamfer_values, outer_male_chamfer_values)
    
    print("\nMann-Whitney U test results (Outer Dataset P2S Distance):")
    print(f"U-statistic: {outer_p2s_mw.statistic:.4f}")
    print(f"p-value: {outer_p2s_mw.pvalue:.4f}")
    print(f"Statistically significant (p < 0.05): {outer_p2s_mw.pvalue < 0.05}")
    
    print("\nMann-Whitney U test results (Outer Dataset Chamfer Distance):")
    print(f"U-statistic: {outer_chamfer_mw.statistic:.4f}")
    print(f"p-value: {outer_chamfer_mw.pvalue:.4f}")
    print(f"Statistically significant (p < 0.05): {outer_chamfer_mw.pvalue < 0.05}")
    
    # Check normality assumptions for outer dataset
    print("\n=== NORMALITY TESTS (OUTER DATASET) ===")
    _, p_outer_female_p2s = stats.shapiro(outer_female_dist_values)
    _, p_outer_male_p2s = stats.shapiro(outer_male_dist_values)
    _, p_outer_female_chamfer = stats.shapiro(outer_female_chamfer_values)
    _, p_outer_male_chamfer = stats.shapiro(outer_male_chamfer_values)
    
    print(f"Shapiro-Wilk test p-values (p < 0.05 indicates non-normal distribution):")
    print(f"Outer Female P2S: {p_outer_female_p2s:.4f} (Normal: {p_outer_female_p2s >= 0.05})")
    print(f"Outer Male P2S: {p_outer_male_p2s:.4f} (Normal: {p_outer_male_p2s >= 0.05})")
    print(f"Outer Female Chamfer: {p_outer_female_chamfer:.4f} (Normal: {p_outer_female_chamfer >= 0.05})")
    print(f"Outer Male Chamfer: {p_outer_male_chamfer:.4f} (Normal: {p_outer_male_chamfer >= 0.05})")
    
    # Create visualizations for outer dataset
    create_visualizations("Outer", outer_female_dist_values, outer_male_dist_values, 
                         outer_female_chamfer_values, outer_male_chamfer_values)
    
    return {
        'inner_p2s_ttest': inner_p2s_ttest,
        'inner_chamfer_ttest': inner_chamfer_ttest,
        'inner_p2s_mw': inner_p2s_mw,
        'inner_chamfer_mw': inner_chamfer_mw,
        'outer_p2s_ttest': outer_p2s_ttest,
        'outer_chamfer_ttest': outer_chamfer_ttest,
        'outer_p2s_mw': outer_p2s_mw,
        'outer_chamfer_mw': outer_chamfer_mw
    }

# Run the statistical tests
results = test_gender_differences()

# Print effect size (Cohen's d) for additional context
def calculate_effect_size():
    # Extract actual data points for inner dataset
    inner_female_dist_values = inner_distance_overall[inner_distance_overall['model_id'].isin(female_ids)]['mean']
    inner_male_dist_values = inner_distance_overall[inner_distance_overall['model_id'].isin(male_ids)]['mean']
    
    inner_female_chamfer_values = inner_chamfer_overall[inner_chamfer_overall['model_id'].isin(female_ids)]['chamfer_distance']
    inner_male_chamfer_values = inner_chamfer_overall[inner_chamfer_overall['model_id'].isin(male_ids)]['chamfer_distance']
    
    # Extract actual data points for outer dataset
    outer_female_dist_values = outer_distance_overall[outer_distance_overall['model_id'].isin(female_ids)]['mean']
    outer_male_dist_values = outer_distance_overall[outer_distance_overall['model_id'].isin(male_ids)]['mean']
    
    outer_female_chamfer_values = outer_chamfer_overall[outer_chamfer_overall['model_id'].isin(female_ids)]['chamfer_distance']
    outer_male_chamfer_values = outer_chamfer_overall[outer_chamfer_overall['model_id'].isin(male_ids)]['chamfer_distance']
    
    print("\n=== EFFECT SIZE (COHEN'S D) ===")
    print("\n--- INNER DATASET ---")
    
    # Calculate Cohen's d for Inner P2S
    mean_diff_inner_p2s = inner_female_dist_values.mean() - inner_male_dist_values.mean()
    pooled_std_inner_p2s = np.sqrt(((len(inner_female_dist_values) - 1) * inner_female_dist_values.std()**2 + 
                               (len(inner_male_dist_values) - 1) * inner_male_dist_values.std()**2) / 
                              (len(inner_female_dist_values) + len(inner_male_dist_values) - 2))
    cohens_d_inner_p2s = mean_diff_inner_p2s / pooled_std_inner_p2s
    
    # Calculate Cohen's d for Inner Chamfer
    mean_diff_inner_chamfer = inner_female_chamfer_values.mean() - inner_male_chamfer_values.mean()
    pooled_std_inner_chamfer = np.sqrt(((len(inner_female_chamfer_values) - 1) * inner_female_chamfer_values.std()**2 + 
                                   (len(inner_male_chamfer_values) - 1) * inner_male_chamfer_values.std()**2) / 
                                  (len(inner_female_chamfer_values) + len(inner_male_chamfer_values) - 2))
    cohens_d_inner_chamfer = mean_diff_inner_chamfer / pooled_std_inner_chamfer
    
    print(f"Inner P2S Distance: {cohens_d_inner_p2s:.4f}")
    print(f"Inner Chamfer Distance: {cohens_d_inner_chamfer:.4f}")
    
    print("\n--- OUTER DATASET ---")
    
    # Calculate Cohen's d for Outer P2S
    mean_diff_outer_p2s = outer_female_dist_values.mean() - outer_male_dist_values.mean()
    pooled_std_outer_p2s = np.sqrt(((len(outer_female_dist_values) - 1) * outer_female_dist_values.std()**2 + 
                               (len(outer_male_dist_values) - 1) * outer_male_dist_values.std()**2) / 
                              (len(outer_female_dist_values) + len(outer_male_dist_values) - 2))
    cohens_d_outer_p2s = mean_diff_outer_p2s / pooled_std_outer_p2s
    
    # Calculate Cohen's d for Outer Chamfer
    mean_diff_outer_chamfer = outer_female_chamfer_values.mean() - outer_male_chamfer_values.mean()
    pooled_std_outer_chamfer = np.sqrt(((len(outer_female_chamfer_values) - 1) * outer_female_chamfer_values.std()**2 + 
                                   (len(outer_male_chamfer_values) - 1) * outer_male_chamfer_values.std()**2) / 
                                  (len(outer_female_chamfer_values) + len(outer_male_chamfer_values) - 2))
    cohens_d_outer_chamfer = mean_diff_outer_chamfer / pooled_std_outer_chamfer
    
    print(f"Outer P2S Distance: {cohens_d_outer_p2s:.4f}")
    print(f"Outer Chamfer Distance: {cohens_d_outer_chamfer:.4f}")
    
    print("\nInterpretation:")
    print("0.2 = small effect, 0.5 = medium effect, 0.8 = large effect")

# Calculate effect size
calculate_effect_size()
```
